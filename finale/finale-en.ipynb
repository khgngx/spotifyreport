**Issue 2** - Format the artists column

```python
import pandas as pd
import ast
df1 = pd.read_csv(r"D:/dataset/messy/music_attributes.csv")
# nh ngha hm format_artists_column  x l ct ngh s trong dataframe
def format_artists_column(df1, col_name='artists'):
    # p dng cho tng gi tr trong ct col_name (mc nh l 'artists')
    # Vi mi gi tr x (chui dng danh sch ngh s)
    df1[col_name] = df1[col_name].apply(
        lambda x: 
            # Chuyn chui x thnh list Python thc s, v d: "['A', 'B']" -> ['A', 'B']
            ast.literal_eval(x)  
    # Ni cc phn t trong list thnh chui, ngn cch bng du phy v khong trng
    ).apply(lambda lst: ", ".join(lst))
    
    # Tr v dataframe  c chnh sa ct artists
    return df1
df1 = format_artists_column(df1, 'artists')
df1.head()
```

**Issue 3** - Column Naming for Clarity and Consistency

```python
# i tn cc ct trong DataFrame df1  r rng hn v phm vi hoc n v o
df1 = df1.rename(columns={
    'valence': 'valence (01)',
    'acousticness': 'acoustic_score (01)',
    'danceability': 'dance_score (01)',
    'energy': 'energy_level (01)',
    'instrumentalness': 'instrumental_score (01)',
    'liveness': 'live_presence (01)',
    'speechiness': 'speech_score (01)',
    'loudness': 'loudness (dB)',
    'tempo': 'tempo (BPM)'
})

df1.head()
```

```python
import pandas as pd
import ast

# c li df2 v lm sch nh cc cell trc
df2 = pd.read_csv(r"D:/dataset/messy/structure_and_popularity.csv")

def format_artists_column(df2, col_name='artists'):
	df2[col_name] = df2[col_name].apply(
		lambda x: ast.literal_eval(x)
	).apply(lambda lst: ", ".join(lst))
	return df2

df2 = format_artists_column(df2)

# i tn ct cho ng nht vi cc bc trc
df2.rename(columns={
	'release_date': 'release_year (YYYY)',
	'year': 'recorded_year (YYYY)',
	'explicit': 'explicit_content (0 = no, 1 = yes)',
	'key': 'musical_key (0-11)',
	'mode': 'mode (1 = major, 0 = minor)',
	'popularity': 'popularity_score (0-100)'
}, inplace=True)

# Merge vi df1  c lm sch
merged_df = pd.merge(df2, df1, on=['artists', 'id', 'name'], how='inner')
merged_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"D:/dataset/messy/data_by_year.csv")
df.head()
def analyze_acousticness_vs_popularity(df):
    # To hnh v vi kch thc 10x6
    plt.figure(figsize=(10, 6))
    # i tn ct cho ng nht
    df.rename(columns={"acoustic_ratio": "acousticness"}, inplace=True)##
    # V biu  scatter vi  trong sut (alpha) l 0.6
    sns.scatterplot(data=df, x='acousticness', y='popularity', alpha=0.6)##
    # t tiu  cho biu 
    plt.title('Relationship between Acousticness and Popularity')
    # t nhn cho trc x
    plt.xlabel('Acousticness (0-1)')
    # t nhn cho trc y
    plt.ylabel('Popularity (0-100)')
    # Thm li vi  trong sut 0.7
    plt.grid(True, alpha=0.7)
    # Hin th biu 
    plt.show()
    # Tnh v in h s tng quan gia acousticness v popularity
    correlation = df['acousticness'].corr(df['popularity'])##
    print(f"Correlation: {correlation:.2f}")

# Gi hm  phn tch d liu t DataFrame df
analyze_acousticness_vs_popularity(df)
```

- Mi tng quan gia cc thuc tnh cu trc bi ht (v d: nhp  (tempo),  ln (loudness), tng (key)) v mc  ph bin l g?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Tnh h s tng quan Pearson gia cc thuc tnh
corr = df[['tempo', 'loudness', 'key', 'popularity']].corr()

print("Correlation matrix:")
print(corr)

# 2. V heatmap  trc quan mi tng quan
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix between Song Attributes and Popularity')
plt.show()
```

```python
# 3. V scatter plots  xem quan h gia tng thuc tnh v popularity
fig, axs = plt.subplots(1, 3, figsize=(18,5))

# Scatter tempo vs popularity
sns.scatterplot(data=df, x='tempo', y='popularity', ax=axs[0])
axs[0].set_title('Tempo vs Popularity')

# Scatter loudness vs popularity
sns.scatterplot(data=df, x='loudness', y='popularity', ax=axs[1])
axs[1].set_title('Loudness vs Popularity')

# Scatter key vs popularity (v key l dng s ri rc, dng stripplot hoc swarmplot)
sns.stripplot(data=df, x='key', y='popularity', ax=axs[2], jitter=True)
axs[2].set_title('Key vs Popularity')

plt.tight_layout()
plt.show()
```
