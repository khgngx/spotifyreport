```python
import pandas as pd

# Tạo hai bảng (dataframes)
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [70000, 80000, 50000, 60000]
})

merged_df = pd.merge(df1, df2, on='ID', how='left')  # 'inner' chỉ giữ giá trị chung

print(merged_df)
```
