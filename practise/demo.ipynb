```python
#Tạo 1 mảng có size  1 000 000 phần tử và tính trung bình cộng
import random
size = 1_000_000
array = [random.uniform(0, 100) for _ in range(size)]  # random.uniform sinh số thực ngẫu nhiên
# Tính trung bình cộng
sum_array = sum(array)
average = sum_array / size
print(f"Tổng các phần tử: {sum_array:.2f}")
print(f"Trung bình cộng: {average:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dữ liệu mẫu
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Vẽ biểu đồ
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Dữ liệu')

# Thêm tiêu đề và nhãn
plt.title("Biểu đồ đường cơ bản")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.legend()
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```
