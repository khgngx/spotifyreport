```python
import pandas as pd

# Đọc dữ liệu
df1 = pd.read_csv(r"D:/dataset/messy/music_attributes.csv")

# ✅ 1. Kiểm tra range và thống kê các cột số
cols_to_check = ['year', 'explicit', 'key', 'mode', 'popularity']

print("=== 📊 RANGE CÁC CỘT SỐ ===\n")
for col in cols_to_check:
    print(f"📌 Cột: {col}")
    print(f"- Min: {df1[col].min()}")
    print(f"- Max: {df1[col].max()}")
    print(f"- Số giá trị duy nhất: {df1[col].nunique()}")
    if df1[col].nunique() <= 20:
        print(f"- Các giá trị duy nhất: {df1[col].unique()}")
    else:
        print("- Các giá trị duy nhất: quá nhiều, không hiển thị.")
    print()

# ✅ 2. Kiểm tra nội dung cột 'artists'
print("=== 🎤 KIỂM TRA CỘT 'artists' ===\n")
print("📌 Một số giá trị mẫu:")
print(df1['artists'].sample(10))

print(f"\n🔢 Số nghệ sĩ duy nhất: {df1['artists'].nunique()}")

print("\n🏆 Top 10 nghệ sĩ xuất hiện nhiều nhất:")
print(df1['artists'].value_counts().head(10))
```
