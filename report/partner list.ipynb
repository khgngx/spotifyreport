```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel(r'D:\dataset\messy\Cus call list.xlsx')
df = df.rename(columns={'CustomerID': 'CustomerID',
                         'First_Name': 'First Name', 
                         'Last_Name': 'Last Name', 
                         'Phone_Number': 'Phone Number', 
                         'Address': 'Address',
                         'Paying Customer': 'Paying Customer', 
                         'Do_Not_Contact': 'Do Not Contact', 
                         'Not_Useful_Column': 'Not Useful Column'})

def cleanPhoneNumbers(PhoneNumber):
    PhoneNumber = str(PhoneNumber) 
    # Check if the value is NaN or in ["n/a", "na"]
    if pd.isna(PhoneNumber) or str(PhoneNumber).lower() in ["n/a", "na"]:
        return "0"
    # Use regex to remove all non-digit characters
    PhoneNumber = re.sub(r"[^\d]", "", PhoneNumber)
    # If the phone number has 10 digits, format it as 3-3-4
    if len(PhoneNumber) == 10:
        return f"{PhoneNumber[:3]}-{PhoneNumber[3:6]}-{PhoneNumber[6:]}"
    # If the phone number has 11 digits and starts with 1, format it as 1-3-3-4
    elif len(PhoneNumber) == 11 and PhoneNumber[0] == "1":
        return f"{PhoneNumber[1:4]}-{PhoneNumber[4:7]}-{PhoneNumber[7:]}"
    else:
        return "0"

def cleanLastName(LastName):
    LastName = str(LastName)
    # Check if the value is NaN or in ["n/a", "na"]
    if pd.isna(LastName) or str(LastName).lower() in ["n/a", "na"]:
        return "0"
    # Loc cac ky tu khong phai la chu cai hoac khoang trang
    LastName = re.sub(r"[^a-zA-Z\s]", "", LastName)
    # Xoa khoang trang thua
    LastName = " ".join(LastName.split()) 
    return LastName

df = df.drop_duplicates()
df['Phone Number'] = df['Phone Number'].apply(cleanPhoneNumbers)
df['Last Name'] = df['Last Name'].apply(cleanLastName)
df["Paying Customer"] = df["Paying Customer"].replace({"Yes": "Y", "No": "N","N/a":"N"})
df["Do Not Contact"] = df["Do Not Contact"].replace({"Yes": "Y", "No": "Y","N/a":"N"})
df["Address"] = df["Address"].replace({"N/a":"No Address"})
df = df.drop(columns=["Not Useful Column"])
df.fillna("N", inplace=True)

df
```
