### Problem 1: Check for Missing or Redundant Data (Missing Values)

```python
import pandas as pd
# Đọc dữ liệu từ file CSV
df = pd.read_csv(r'D:\dataset\messy\calendar1.csv')
# Bước 1: Chuyển đổi cột 'price' sang kiểu số (loại bỏ dấu '$' và ',' nếu có)
df['price'] = df['price'].astype(str).str.replace(r'[\$,]', '', regex=True)
# Bước 2: Xử lý các giá trị bị thiếu (NaN) trong 'price'
df['price'] = pd.to_numeric(df['price'], errors='coerce')  # Chuyển tất cả về số (float)
df.loc[df['available'] == 'f', 'price'] = 0  # Nếu không khả dụng, đặt giá là 0
df['price'] = df['price'].fillna(df['price'].mean()) # Điền giá trị trung bình cho phần còn lại
# Kiểm tra lại sau khi xử lý
df['available'] = df['available'].replace({'t': 'Yes', 'f': 'No'}).fillna(0)
# Hiển thị 5 dòng đầu tiên để kiểm tra kết quả
df.head()
```

```python
import pandas as pd

# Kiểm tra số lượng giá trị NaN
print(df1.isnull().sum())

# Xóa cột 'square_feet' nếu có quá nhiều NaN
df1.drop(columns=['square_feet'], inplace=True)

# Xóa các dòng có quá nhiều giá trị NaN
df1.dropna(thresh=df1.shape[1] * 0.5, inplace=True)

# Điền giá trị NaN cho cột kiểu text
df1.loc[:, 'summary'] = df1['summary'].fillna("No summary available")
df1.loc[:, 'space'] = df1['space'].fillna("No space description")

# Điền giá trị NaN cho cột số bằng mean hoặc mode
df1.loc[:, 'bathrooms'] = df1['bathrooms'].fillna(df1['bathrooms'].mean())
df1.loc[:, 'bedrooms'] = df1['bedrooms'].fillna(df1['bedrooms'].mode()[0])

# Xử lý cột price (loại bỏ ký tự đặc biệt, chuyển sang số)
df1.loc[:, 'price'] = df1['price'].replace(r'[\$,]', '', regex=True).astype(float)
df1.loc[:, 'price'] = df1['price'].fillna(df1['price'].mean())

# Kiểm tra lại sau khi xử lý
print(df1.isnull().sum())
```

### **Problem 2: Remove Unnecessary Columns and Rename Column Name**


```python
import numpy as np
columns_to_drop = [
    'square_feet',  # Quá nhiều giá trị NaN
    'country_code',  # Thông tin trùng lặp với 'country'
    'description',  # Thông tin gần như trùng với 'summary'
    'host_name',  # Không cần thiết nếu đã có 'host_id'
]
df1.rename(columns={
    'id': 'ID',
    'name': 'Listing Name',
    'summary': 'Listing Summary',
    'space': 'Listing Space',
    'host_id': 'Owner ID',
    'neighbourhood_group_cleansed': 'Region',
    'city': 'City',
    'country': 'Country',
    'property_type': 'Property Type',
    'room_type': 'Room Type',
    'accommodates': 'Guest Capacity',
    'bathrooms': 'Bathrooms',
    'bedrooms': 'Bedrooms',
    'beds': 'Beds',
    'bed_type': 'Bed Type',
    'amenities': 'Amenities',
    'price': 'Price'
}, inplace=True)
# Kiểm tra lại DataFrame sau khi đổi tên
print(df1.head())

# Chỉ giữ lại những cột có trong DataFrame trước khi xóa
existing_columns = [col for col in columns_to_drop if col in df1.columns]

# Nếu có cột hợp lệ, thực hiện xóa
if existing_columns:
    df1.drop(columns=existing_columns, inplace=True)
else:
    print("")

# 2. Kiểm tra tương quan giữa các cột số và loại bỏ cột dư thừa
numeric_df1 = df1.select_dtypes(include=[np.number])
corr_matrix = numeric_df1.corr()
df1.head()
```

 ### **Problem 3: Standalize Data**


```python
import pandas as pd
import re
# Chuẩn hóa giá cột 'Price' (loại bỏ ký tự '$' và chuyển về số)
df['price'] = df['price'].replace('[\\$,]', '', regex=True).astype(float)
# Chuẩn hóa cột 'Amenities' (chuyển về chữ thường, bỏ dấu ngoặc vuông)
df1['Amenities'] = df1['Amenities'].astype(str).str.lower().str.replace(r'[\[\]"]', '', regex=True)
# Xử lý giá trị thiếu (NaN) trong các cột số
df1['Bathrooms'] = df1['Bathrooms'].fillna(df1['Bathrooms'].mean())
df1['Bedrooms'] = df1['Bedrooms'].fillna(df1['Bedrooms'].mode()[0])
df1['Beds'] = df1['Beds'].fillna(df1['Beds'].mode()[0])
df1.head()
```

 ### **Problem 4: Handle Anomalous Data**


```python
import pandas as pd
# Xác định phạm vi IQR cho cột 'Price'
Q1 = df1['Price'].quantile(0.25)
Q3 = df1['Price'].quantile(0.75)
IQR = Q3 - Q1
# Giới hạn hợp lý cho giá phòng
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Loại bỏ các dòng có giá trị ngoại lai
df1 = df1[(df1['Price'] >= lower_bound) & (df1['Price'] <= upper_bound)]
# Loại bỏ những dòng có số giường, số phòng ngủ hoặc phòng tắm âm hoặc bằng 0
df1 = df1[(df1['Bathrooms'] > 0) & (df1['Bedrooms'] > 0) & (df1['Beds'] > 0)]
# Nếu Room Type là 'Entire Home/Apt' thì số Beds ít nhất phải là 1
df1.loc[(df1['Room Type'] == 'Entire Home/Apt') & (df1['Beds'] == 0), 'Beds'] = 1
# Hiển thị dữ liệu sau khi xử lý
df1.head()
```

### **Merge Dataset**


```python
# Đổi tên cột 'listing_id' trong df để khớp với 'ID' trong df1
df.rename(columns={'listing_id': 'ID'}, inplace=True)
# Chuẩn hóa cột 'price' trong df (bỏ dấu $ và chuyển về dạng float)
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
# Merge hai bảng dựa trên 'ID'
merged_df = df1.merge(df, on='ID', how='inner')  # 'inner' để chỉ giữ ID trùng khớp
# Hiển thị kết quả
print(merged_df.head())
```
