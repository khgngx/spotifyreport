```python
import numpy as np

def input_data(n):
    return np.random.randint(0, 100, n)

def print_array(A):
    print("Array A:", A)

def compute_sum(A):
    return np.sum(A)

def compute_average(A):
    return np.mean(A)

def find_x(A, x):
    return np.where(A == x)[0]

def insert_element(A, i, x):
    return np.insert(A, i, x)

def remove_element(A, i):
    return np.delete(A, i)

# Example usage
n = int(input("Enter the size of the array: "))
A = input_data(n)
print_array(A)

print("Sum of the array:", compute_sum(A))
print("Average of the array:", compute_average(A))

x = int(input("Enter the value to find: "))
positions = find_x(A, x)
if len(positions) > 0:
    print(f"Value {x} found at positions: {positions}")
else:
    print(f"Value {x} not found in the array.")

i = int(input("Enter the position to insert a new element: "))
x = int(input("Enter the value to insert: "))
A = insert_element(A, i, x)
print("Array after insertion:")
print_array(A)

i = int(input("Enter the position to remove an element: "))
A = remove_element(A, i)
print("Array after removal:")
print_array(A)
```
