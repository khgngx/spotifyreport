3 - Count all odd numbers

```python
def sumOddNumbers(A):
    # Tinh tong cac phan tu le trong mang A
    return sum(i for i in A if i % 2 != 0)
```

4 - Compute sum of the array

```python
def computeSum(A):
    return sum(A)
```

5 - Compute average of the array.

```python
def computeAverage(A):
    # Tinh trung binh cong cua mang A
    return sum(A)/len(A) if A else 0
    # Neu A khong co phan tu nao thi tra ve 0
```

6 - Count elements in range a to b

```python
def countElement(A,a,b):
    # Tinh gia tri cac phan tu trong mang A thuoc [a,b]
    return sum(i for i in A if a <= i <= b)
```

```python
import random

n = int(input("Nhập kích thước của mảng: "))
A = [random.randint(0, 100) for _ in range(n)]

def computeSum(A):
    return sum(A)

def computeAverage(A):
    return sum(A) / len(A) if len(A) > 0 else 0

def sumOddNumbers(A):
    return sum(x for x in A if x % 2 != 0)

def countElement(A, a, b):
    return sum(1 for x in A if a <= x <= b)

def insertElement(A, k, x):
    return A[:k] + [x] + A[k:]

def removeElement(A, k):
    return A[:k] + A[k+1:]

print("Mảng A:", A)
k = int(input("Nhập vị trí k: "))
x = int(input("Nhập giá trị x: "))
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
print("Tổng các phần tử trong mảng A:", computeSum(A))
print("Trung bình cộng các phần tử trong mảng A:", computeAverage(A))
print("Tổng các phần tử lẻ trong mảng A:", sumOddNumbers(A))
print("Số phần tử trong mảng A thuộc [a, b]:", countElement(A, a, b))
print("Mảng A sau khi chèn phần tử x vào vị trí k:", insertElement(A, k, x))
print("Mảng A sau khi xóa phần tử ở vị trí k:", removeElement(A, k))
```
